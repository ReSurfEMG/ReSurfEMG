EMGs are a type of file that can be read into your computer. The file is a strucured sets of data. The data includes many data types, but the signal is a group of arrays (one for each lead). You can think of arrays as a special kind of list. We can use already written groups of functions in libraries to deal with arrays, and graph them. We will also use a libraries to help us make fake (synthetic) EMGs

```python
# import libraries

import matplotlib.pyplot as plt
import numpy as np
import math
```

```python


x = np.linspace(0, 20, 10000)
raised_sin = np.sin(x*6/60*2*math.pi)**2
synth_emg1 = raised_sin * np.random.randn((len(x)))+ 0.1 * np.random.randn(len(x))
#longt = 
plt.plot(x,synth_emg1 )#'.^2 .* randn(length(t),1) + 0.1 * randn(length(t), 1))
plt.xlabel('Time (sec)')
plt.ylabel('EMG (a.u.)')
```
